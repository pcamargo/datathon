## Contexto do Problema

O objetivo desta análise é compreender os fatores que influenciam o desenvolvimento dos alunos ao longo do programa Passos Mágicos, utilizando indicadores acadêmicos, comportamentais e psicossociais.

Buscamos identificar padrões de evolução, antecipar riscos de defasagem e apoiar decisões mais assertivas por meio de dados.

````python
# Bibliotecas principais
import pandas as pd
import numpy as np

# Visualização
import matplotlib.pyplot as plt
import seaborn as sns

# Estatística
from scipy.stats import spearmanr

# Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import roc_curve

# Configurações
plt.style.use("seaborn-v0_8")
pd.set_option("display.max_columns", 100)
````

````python
def read_data(path):
    abas = ["PEDE2022", "PEDE2023", "PEDE2024"]

    dfs = []

    # Leitura e marcação do ano
    for aba in abas:
        try:
            df_temp = pd.read_excel(path, sheet_name=aba)
            df_temp["ANO"] = aba
            dfs.append(df_temp)
        except Exception as e:
            print(f"Erro ao ler a aba {aba}: {e}")

    if not dfs:
        return pd.DataFrame()

    # Identificar todas as colunas únicas presentes em todos os arquivos
    # Usando uma lista ordenada para manter a consistência, se desejar
    todas_colunas = sorted(list(set().union(*(d.columns for d in dfs))))

    dfs_padronizados = []

    for d in dfs:
        # Criar as colunas faltantes preenchidas com NaN
        d_padronizado = d.reindex(columns=todas_colunas)

        # Filtrando DFs não vazios de dados e colunas
        if not d_padronizado.empty or len(d_padronizado.columns) > 0:
            dfs_padronizados.append(d_padronizado)

    # Concatenação segura
    if not dfs_padronizados:
        return pd.DataFrame(columns=todas_colunas)

    df = pd.concat(dfs_padronizados, ignore_index=True)
    return df
````

````python
# criar variável de risco com base em critérios definidos
df["risco"] = np.where(
    (df["ian"] > 4) | (df["ida"] < df["ida"].median()),
    1,
    0
)
````

````python
# verificar a distribuição do risco
df["risco"].value_counts(normalize=True)
````

````python
# definir as features
feature_cols = indicadores
````

````python
# separar as features e a variável alvo
X = df[feature_cols]
y = df["risco"]
````

````python
# padronizar as features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
# separar os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
````

````python
# treinar um modelo de regressão logística
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
````

````python
# fazer previsões
y_pred = log_model.predict(X_test)
y_prob = log_model.predict_proba(X_test)[:, 1]
````

````python
# avaliar o modelo
print(classification_report(y_test, y_pred, zero_division=0))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
````

````python
# treinar um modelo de Random Forest
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight="balanced"
)
````

````python
# treinar o modelo
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]
````

````python
# avaliar o modelo de Random Forest
print(classification_report(y_test, y_pred_rf))
print("ROC AUC RF:", roc_auc_score(y_test, y_prob_rf))
````

````python
# plotar a importância das variáveis no modelo de Random Forest
importances = pd.Series(
    rf_model.feature_importances_,
    index=feature_cols
).sort_values(ascending=False)
````

````python
import shap
````

````python
# criar o objeto SHAP explainer
X_shap = X_train.copy()
explainer = shap.TreeExplainer(rf_model)
# calcular os valores SHAP para o conjunto de teste
shap_values = explainer.shap_values(X_shap)

# plotar o gráfico de resumo SHAP para visualizar as importâncias das variáveis
shap.summary_plot(shap_values, X_shap, plot_type="bar")
plt.show()
````

**Interpretando a interdependência entre as variáveis**:

**Distribuição de Impacto**: Note como os pontos se agrupam em "nuvens". Onde a nuvem é mais larga, há uma maior densidade de alunos com aquele perfil de impacto no risco.

**Feature 0 vs Feature 1**: Quando os pontos mudam de cor (de azul para vermelho) ao longo do eixo horizontal, o modelo está nos dizendo que o valor daquela variável altera drasticamente a predição.

**Sensibilidade**: Se houver uma separação clara entre as cores, o modelo está bem calibrado. Se houver muita mistura, pode indicar que o modelo precisa de mais dados ou de uma técnica de feature engineering para separar melhor esses perfis

````python
# calcular os valores SHAP para o conjunto de teste
shap_values_test = explainer(X_test)
````

````python
# plotar o gráfico de waterfall para visualizar as contribuições SHAP para a previsão do risco para a instância idx

idx = 0
exp_aluno = shap_values_test[idx]
shap.plots.waterfall(
    exp_aluno[:, 1]
)
````

**Interpretação**

**O Valor Base**: O modelo começa com um valor esperado (a média).

**Contribuições Positivas (Vermelho)**: Variáveis como ipv (+0.05) e ips (+0.04) estão empurrando este aluno específico para uma zona de maior risco.

**Contribuições Negativas (Azul)**: O ieg (-0.07) e o ida (-0.03) estão atuando como "protetores", puxando o score de risco para baixo.

**Resultado Final ($f(x) = 0.482$)**: É a soma algébrica dessas forças. Este aluno está quase no "limiar" de 0.5 (se esse for o seu cutoff para risco alto).

````python
explicacao_aluno = pd.DataFrame({
    "Indicador": features_modelo,
    "Impacto_SHAP": shap_values_test[idx, :, 1].values
}).sort_values(by="Impacto_SHAP", ascending=False)

explicacao_aluno
````

A matriz de confusão para o modelo Random Forest (`cm`):
```
[[539, 143],
 [ 29, 198]]
```

Vamos interpretar cada quadrante no contexto da predição de risco educacional para o programa Passos Mágicos, onde:
- **Classe 0:** Aluno **não** em risco (ou risco baixo/médio)
- **Classe 1:** Aluno **em risco** (ou risco alto/muito alto)

1.  **Valores da Matriz de Confusão:**
    *   **Verdadeiros Negativos (TN): 539** (Superior Esquerdo)
        *   O modelo previu corretamente que 539 alunos **não** estavam em risco, e eles de fato não estavam.
    *   **Falsos Positivos (FP): 143** (Superior Direito)
        *   O modelo previu que 143 alunos estavam em risco, mas eles, na verdade, não estavam (ou seja, foram alarmes falsos).
    *   **Falsos Negativos (FN): 29** (Inferior Esquerdo)
        *   O modelo previu que 29 alunos **não** estavam em risco, mas eles, na verdade, estavam (ou seja, o modelo "perdeu" alunos em risco).
    *   **Verdadeiros Positivos (TP): 198** (Inferior Direito)
        *   O modelo previu corretamente que 198 alunos estavam em risco, e eles de fato estavam.

2.  **Significado de cada métrica no contexto do programa:**
    *   **Verdadeiros Negativos (539):** Representam os alunos bem classificados que não precisariam de intervenção especial e que o programa não direcionaria recursos de forma desnecessária. É um bom indicativo da eficiência do modelo em evitar intervenções equivocadas.
    *   **Falsos Positivos (143):** São os alunos que o modelo identificou como em risco, mas que não estavam. No contexto educacional, isso significa que 143 alunos poderiam receber atenção extra ou serem incluídos em programas de apoio desnecessariamente. Embora não seja tão crítico quanto um Falso Negativo, pode levar ao uso ineficiente de recursos e à estigmatização de alunos.
    *   **Falsos Negativos (29):** São os alunos mais preocupantes. O modelo falhou em identificar 29 alunos que estavam em risco. Para o programa Passos Mágicos, isso significa que esses alunos, que realmente precisavam de apoio, não seriam sinalizados e poderiam continuar com defasagem ou ter seu desempenho prejudicado sem a intervenção necessária. Essa é uma falha crítica, pois o objetivo é justamente identificar e apoiar esses alunos.
    *   **Verdadeiros Positivos (198):** São os alunos corretamente identificados como em risco, permitindo que o programa Passos Mágicos atue preventivamente ou com intervenções direcionadas para ajudá-los. Um alto número de Verdadeiros Positivos demonstra a capacidade do modelo de cumprir seu objetivo principal.

3.  **Balanceamento entre Tipos de Erro:**
    Para o programa Passos Mágicos, os **Falsos Negativos (29)** são mais críticos. Deixar de identificar um aluno em risco (FN) pode ter consequências mais graves para o desenvolvimento individual do aluno do que identificar erroneamente um aluno como em risco (FP). Um Falso Positivo pode levar a um uso ineficiente de recursos, mas um Falso Negativo implica que um aluno que necessita de ajuda não a receberá, potencialmente agravando sua situação acadêmica e psicossocial. O programa busca evitar a defasagem, portanto, perder alunos em risco é um erro mais custoso.
